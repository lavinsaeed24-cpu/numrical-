```python
def bisection(a, b, tol=1e-6, max_iter=50):
    # Initial check
    if f(a) * f(b) > 0:
        print("Bisection method fails: f(a) and f(b) must have opposite signs.")
        return None

    # Print header
    print(f"{'Iter':>4} | {'a':>10} | {'b':>10} | {'c':>10} | {'f(c)':>12}")
    print("-" * 55)

    # Main iteration loop
    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        print(f"{i+1:>4} | {a:>10.6f} | {b:>10.6f} | {c:>10.6f} | {fc:>12.6e}")

        # Check convergence
        if abs(fc) < tol or (b - a) / 2 < tol:
            print(f"\nApproximate root = {c:.6f} after {i+1} iterations.")
            return c

        # Update interval
        if f(a) * fc < 0:
            b = c
        else:
            a = c

    # If max iterations reached
    print("Max iterations reached.")
    return (a + b) / 2
```

```python
# Set the interval [a, b]
a = 2
b = 3
print(f"Finding root in interval [{a}, {b}]")
```

```python
root = bisection(a, b)
```

```python
if root is not None:
    print(f"\n" + "="*50)
    print(f"Final Estimated root: {root:.6f}")
    print(f"Verification: f({root:.6f}) = {f(root):.6e}")
    print("="*50)
```
