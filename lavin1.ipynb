```python
def f(x):
    return x**3 - 4*x - 9

def bisection(a, b, tol=1e-6, max_iter=50):
    if f(a) * f(b) > 0:
        print("Bisection method fails: f(a) and f(b) must have opposite signs.")
        return None
    print(f"{'Iter':>4} | {'a':>10} | {'b':>10} | {'c':>10} | {'f(c)':>12}")
    print("-" * 55)
    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        print(f"{i+1:>4} | {a:>10.6f} | {b:>10.6f} | {c:>10.6f} | {fc:>12.6e}")
        if abs(fc) < tol or (b - a) / 2 < tol:
            print(f"\nApproximate root = {c:.6f} after {i+1} iterations.")
            return c
        if f(a) * fc < 0:
            b = c
        else:
            a = c
    print("Max iterations reached.")
    return (a + b) / 2

root = bisection(2, 3)
print(f"\nEstimated root: {root:.6f}")
```
